```python
import numpy as np
import pandas as pd
import re
import itertools
#because Kong Pao Chicken's Recipe name has space in the end, may cause errors 
cookbook = pd.read_csv("recipes.csv",converters ={"Recipe":str.strip})
#fix a spelling error
cookbook.iloc[46,2] = "avocado"
cookbook
```

```python
# since the unit of cooking time is different, we want to change all those into minutes
# create change_time() function to change all the time to minutes 
def change_time(time):
    total_mins = 0
    # use regular expression to deal time
    time = re.split('(\d+)',time)[1:]
    for i in range(0, len(time)-1): 
        # check whether hr
        if time[i].isnumeric() and time[i+1][1] == "h": 
            total_mins += int(time[i]) * 60
        #  check minutes 
        if time[i].isnumeric() and time[i+1][1] == "m": 
            total_mins += int(time[i]) 
    return total_mins
```
