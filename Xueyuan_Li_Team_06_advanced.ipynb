## Question 7: Nutrition Matters

<br> 

*Your user wants to know how to better control their caloric intake by looking at the total calories of a dish or dishes*

<br>

1）Write a function `total_calories` that provides the total calories of the recipe(s) selected by the users. The user will need to specify the recipe(s) they want and the function will return the summative calories of these recipes. Summative calories are calculated by the number of servings each recipe serves times the calories per serving. If the user inputs nothing, print out a message to let them know.<br>
<br>
**Parameters:**<br>
`recipes`: a list of recipes selected by users<br>
**Return:** Total calories of these recipes. <br>
**Example Input:** `["Bulgogi", "Miso Soup", "Paratha"]` <br>
**Example Output:**`4464` <br>

2）Write a function `calories_recipes` that provides different combinations of recipes that meet the user's requirement on the calories intake. The user will need to specify the total caloric intake and the number of recipes. The function should list all the possible combinations of recipes with their summative calories equal to or less than the target caloric intake. If none of the recipe combinations meet the requirement, print out a message to let the user know.<br>
<br>
**Parameters:**<br>
`calories`: an integer indicates the total calries the user intake<br>
`num_recipes`: an interger indicates the number of recipes <br>
**Return:** A list of possible combinations of recipes. Each combination of recipes should as be a list. <br>
**Example Input:** `1500, 2` <br>
**Example Output:**`[["Grilled Cheese Sandwich", "Miso Soup"], ["Gambas al Ajillo", "Miso Soup"], ["Miso Soup", "Guacamole"]]` <br>

3) Implement a user interaction interface for the above two functions. Ask the users to enter their own inputs, and return approriate results or warnings to the user. 

```python
def total_calories(recipes):
    # Edit here
    #recipes=input("").split(',')
    #(1) remove blank place for each recipe you input
    a=[]
    for x in recipes:      
        x=" ".join(x.split())
        a.append(x)
    
    #(2)remove blank place between recipes you input
    recipes = [x.strip(' ') for x in a]
    recipe=[]
    
    #(3)remove number in each recipe
    for i in recipes:
        #print(i)
        i=''.join(filter(lambda x: not x.isdigit(), i))
        
        recipe.append(i)   
    print("The Recipes you want to search are: ", recipe,"\n") 
    
    #(4) case insensitive
    recipe_lower = [word.lower() for word in recipe]
    recipes_lower=[word.lower() for word in cookbook['Recipe']]
    
    if set(recipe_lower ).issubset(set(recipes_lower )):
        print("The Recipes you enter are in the cookbook")
        
        #(5)get each recipe's calory
        calories=[]
        if len(recipe)== 0:
            print("Please input the recipes name that you want to search!")
        else:
            for i in recipe:
                df = cookbook['Recipe'].str.match(i, na=False, case=False)
                index=df.index[df == True].tolist()
                select_df=cookbook.loc[index[0],]
                cal=select_df['Calories per Serving']
                calories.append(cal)
        #(6)get total calories
        print("total calories: ", sum(calories))
    
    else:
        print("The Recipes you enter are not in the cookbook!")    

#recipes=input("").split(',') 
#total_calories(recipes)   
#Grilled Cheese Sandwich,Peking Duck
```

```python
from itertools import permutations

def calories_recipes(calories, num_recipes):
    # Edit here
    #calories=int(input("Total calories: "))
    #num_recipes=int(input("Total number of recipes you want: "))
    
    df=cookbook[['Recipe', 'Calories per Serving']]
    df = df.drop_duplicates()
    cal=list(df['Calories per Serving'])
    rec=list(df['Recipe'])
    
    #Add total calories for each recipe
    comb_cal = list(permutations(cal, num_recipes))
    comb_rec= list(permutations(rec, num_recipes)) 
    Total_cal=[]
    for i in comb_cal:
        Total_cal.append(sum(i))  
        
    #Make a dictionary to save as {recipe:calories}
    dict={'recipe':comb_rec, 'calories':comb_cal, 'Total_cal': Total_cal}
    cal_rec=pd.DataFrame(dict)
    
    #Select the recipe which no more than total calories
    for i in cal_rec:
        cal_rec=cal_rec[cal_rec.Total_cal<=calories]
        
    #if no matched recipes
    if len(cal_rec)==0:
        print("No matched Recipes!")

    #print("All the possible combinations of recipes: ","\n")
    #print(cal_rec['recipe'].to_string(index=False),"\n")
    result=[]
    for j in cal_rec['recipe']:
        result.append(str(j))
    return result

#calories=int(input("Total calories: "))
#num_recipes=int(input("Total number of recipes you want: "))
#calories_recipes(calories, num_recipes)
```

```python
#(3) function2
calories=int(input("Total calories: "))
num_recipes=int(input("Total number of recipes you want: "))
calories_recipes(calories, num_recipes)
```

```python
# GUI
from tkinter import *
class MyWindow:
    def __init__(self, win):
        self.lbl1=Label(win, text='Total calories', width=10 )
        self.lbl2=Label(win, text='Recipes Number ', width=11)
        self.btn1 = Button(win, text='Search')
        self.lbl3=Label(win, text='Combinations')
        self.t1=Entry(bd=3)
        self.t2=Entry()
        self.txt3=Text(win, height=30, width=60 ,bg = "light yellow")
        
        
        self.lbl1.place(x=100, y=50)
        self.t1.place(x=200, y=50)
        
        self.lbl2.place(x=100, y=100)
        self.t2.place(x=200, y=100)
        
        self.b1=Button(win, text='Search', command=self.add)
        self.b1.place(x=100, y=150)
        
        
        self.lbl3.place(x=100, y=200)
        self.txt3.place(x=200, y=200)
        #self.txt3.pack()
        
    def add(self):
        calories=int(self.t1.get())
        num_recipes=int(self.t2.get())
        result=calories_recipes(calories, num_recipes)
        for i in result:
            self.txt3.insert(END, str(i)+"\n")


window=Tk()
mywin=MyWindow(window)
window.title('My Recipe')
window.geometry('700x800+10+10')
window.mainloop()
```
