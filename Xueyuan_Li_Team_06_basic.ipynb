```python
import re
import string
def recipe_info(recipes, cookbook):
    # Edit here
    #recipes=input("What recipe you want to search? ")
    
    #add quantity and ingredient as a new column
    col= cookbook["Quantity"]+" "+cookbook["Ingredient"]
    cookbook["new_list"]=col

    #recipes=" ".join(recipes.split())
    if len(recipes)== 0:
        print("Please input the recipe name that you want to search!")
    else:
        
        #insensitive to blank spaces and numbers, symbols
        result1=re.sub(r'[^\w]', ' ', recipes)
        result = ''.join([i for i in result1 if not i.isdigit()])
        result=" ".join(result.split())

        # case insensitive, judge if the recipe you input contains in the cookbook
        df2 = cookbook['Recipe'].str.fullmatch(result, na=False, case=False)
        # if contain in the cookbook, find the index
        index=df2.index[df2 == True].tolist()
        select_df=cookbook.loc[index,]

        #if the recipe name is found in cookbook
        if len(select_df) != 0:

            print("Name: ",select_df["Recipe"].to_list()[0],"\n")
            print("Recipe:")
            for i in select_df["new_list"]:
                print(i)
            print("\nCook time: ",select_df["Cook Time"].to_list()[0],"\n")
            print("Calories per Serving:",select_df["Calories per Serving"].to_list()[0])
            
        #if the recipe name is not found in cookbook
        else:
            print("The recipe name you input is not included, please re-input!")

            
#recipes=input("What recipe you want to search? ")            
#recipe_info(recipes, cookbook)
```
